```python
import pandas as pd
import matplotlib.pyplot as plt
path = '/content/sample_data/2017_jun_final - 2017_jun_final.csv'
```

```python
df = pd.read_csv(path)
```

```python
print(df.head())
```

```python
print(df.shape) #size of table
```

```python
print(df.dtypes)
```

```python
missing_values = df.isnull().sum()
total_values = df.shape[0]
missing_rate = (missing_values / total_values) * 100
print("\nЧастка пропусків в кожному стовпці y %:")
print(missing_rate)
plt.figure(figsize=(24, 6))
plt.bar(missing_rate.index, missing_rate.values, color='skyblue')
plt.xlabel('Стовпці')
plt.ylabel('Частка пропусків %')
plt.title('Частка пропусків у кожному стовпці')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
lang_index = df.columns.get_loc("Мова програмування")
df_clean = df.dropna(axis=1)
df_clean.insert(lang_index, "Мова програмування", df["Мова програмування"])
print(df_clean.head())
```

```python
missing_values_cleaned = df_clean.isnull().sum()
total_values_cleaned = df_clean.shape[0]
missing_ratio_cleaned = (missing_values_cleaned / total_values_cleaned) * 100
print("\nЧастка пропусків після видалення стовпців (%):")
print(missing_ratio_cleaned)
```

```python
df_clean = df_clean.dropna()
```

```python
print("\nНовий розмір таблицi: ", df_clean.shape)
```

```python
python_data = df_clean[df_clean['Мова програмування'] == 'Python'].copy()
python_data.head(15)
```

```python
print("\nРозмір python_data:", python_data.shape)
```

```python
grouped_data = python_data.groupby('Посада')
```

```python
avg_data = grouped_data.agg({'Зарплата на місяць': ['min', 'max']})
print("\nАгреговані дані за зарплатою:")
print(avg_data)
```

```python
group_avg = python_data.groupby("Посада")["Зарплата на місяць"].mean()

def fill_avg_salary(row):
    return group_avg[row["Посада"]]

python_data["avg"] = python_data.apply(fill_avg_salary, axis=1)

python_data.head()
```

```python
print("\nОписова статистика для стовпчика avg:")
print(python_data['avg'].describe())

plt.figure(figsize=(8, 6))
plt.hist(python_data['avg'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Зарплата на місяць (avg)')
plt.ylabel('Частота')
plt.title('Розподіл середньої зарплати на місяць')
plt.grid(True)
plt.show()
```
